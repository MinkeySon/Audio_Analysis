# 사이즈 유지하면서 필터링 하기 (Where 사용하기)

```python
fb_likes = movie['actor_1_facebook_likes'].dropna()
fb_likes.head()
```

```python
fb_likes.describe()
```

```python
import matplotlib.pyplot as plt
```

```python
c_high  = fb_likes < 20_000
c_high.mean().round(2)
```

```python
fig, ax = plt.subplots(figsize = (10,10))
fb_likes.plot(kind = 'hist')
```

```python
fb_likes
```

```python
fb_likes.where(c_high) # 2만보다 작으면 True, 2보다 크면 Nan 값으로 전환
```

```python
fb_likes.where(c_high,20_000).head()
```

```python
c_low = fb_likes > 300
fb_likes_cap = fb_likes.where(c_high, 20_000).where(c_low,300)
fb_likes_cap.head()
```

```python
fb_likes_cap.plot(kind='hist')
```

이렇게 되면 엄청 큰 값과 엄청 작은 값이 조정이 되면서 그래프가 조정됨
이를 clipping 이라고 함.

```python
fb_likes_cap2 = fb_likes.clip(300,20_000) # fb_likes_cap = fb_likes.where(c_high, 20_000).where(c_low,300) 똑같은 결과
#한 쪽만 할 때는 clip_lower, clip_upper 사용
fb_likes_cap2.equals(fb_likes_cap)
```

```python
c1 = movie.title_year >= 2010
c2 = movie.title_year.isna()
c_all = c1 | c2
movie[c_all]
```

```python
movie.mask(c_all) #title_year가 2010년 이상인 것과 nan인 값은 False 나머지는 냅둠
```

```python
movie_mask = movie.mask(c_all).dropna(how ='all') #how 가 all 일때 모든 값이 na일 경우 삭제 / any 이면 한개라도 있으면 삭제
movie_mask
```

```python
movie_bool = movie[movie.title_year < 2010]
```

```python
print(movie_mask.equals(movie_bool))
print(movie_mask.shape, movie_bool.shape)
```

```python
movie_mask.dtypes == movie_bool.dtypes
```

**float는 3가지 경우가 있음**
1. 진짜 소수인 경우
2. 소수인데 na인 경우
3. 정수인데 na 인경우

```python
from pandas.testing import assert_frame_equal
assert_frame_equal(movie_bool, movie_mask, check_dtype=False)
```
