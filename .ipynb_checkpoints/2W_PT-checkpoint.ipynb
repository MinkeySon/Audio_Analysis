## 1.3. 데이터 형식

결측치를 표시할 수 있는지 여부에 따라 달라집니다.

* float : 부동소수점 (결측치 O) 
* int : 정수 (결측치 X)
* Int64 : 정수 (결측치 O)
* object : 문자열
* category : 범주형 (결측치 O)
* bool : 불리언 (결측치 X)
* boolean : 불리언 (결측치 O)
* datetime64 : 날짜 (결측치 O)

```python
colleges = pd.read_csv('data/college.csv')
colleges.head()
```

```python
colleges.info()
```

```python
colleges = pd.read_csv('data/college.csv', index_col='INSTNM')
# index_col 에서는 어떤 열을 인덱스로 사용할지 지정할 수 있습니다.
# INSTNM 은 학교 이름입니다.
# 학교 이름으로 index 하기 때문에 index 시 활용도 좋음

college_ugds = colleges.filter(like='UGDS_')
college_ugds.head()
```

```python
college_ugds.info()
```

info 메소드 사용해 데이터 형태 확인

```python
college_ugds + 1
```

```python
round(0.5)
```

파이썬은 중간에서 반올림 할 시에 가장 가까운 짝수로 반올림해줌

```python
#여기에 반올림을 적용해봅시다.
#UGDS_BLACK 을 잘 보세요
#반올림 안됨

college_ugds.filter(like='Northwest-Shoals', axis=0).T.round(2)
```

```python
# 이를 막기 위해 작은 수를 더해줍시다.
# add함수를 쓰면 chain 을 깨지 않고 만들 수 있죠!

college_ugds.filter(like='Northwest-Shoals', axis=0).T.add(1e-6).round(2)
```
