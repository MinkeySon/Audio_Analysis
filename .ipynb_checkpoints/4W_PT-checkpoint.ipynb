# 3. 최대 중 최소, 최소 중 최대 - nsmallest, nlargest

"상위 50개 도시 중 가장 물가가 저렴한 5개" 같이 여러 조건을 적용한 데이터가 필요할 수 있습니다.

sort_values 를 여러 번 사용해서 가져올 수도 있겠지만, 이를 빠르게 해주는 메서드가 있습니다.

바로 nlargest 와 nsmallest 입니다.

코드를 보면서 알아봅시다.

```python
movie = pd.read_csv('data/movie.csv')
movie2 = movie[['movie_title', 'imdb_score', 'budget']]
movie2.head()
```

```python
#IMDB 평점이 가장 높은 100개를 먼저 골라보겠습니다.
#nlargest(가져올 개수, 기준이 되는 컬럼들) 로 사용할 수 있습니다.
# 만약 여러 컬럼을 입력할 경우, 동점자의 순서를 가려줍니다.

movie2.nlargest(100, 'imdb_score').head()
```

-> 상위 5위

```python
movie2.sort_values('imdb_score',ascending=False)
```

```python
# 여기서 title_year를 기준으로 정렬해봅시다. sort_values 를 활용합니다.
# 가장 오래된 영화부터 나오겠죠?

(
movie[['movie_title', 'title_year', 'imdb_score']]
    .sort_values('title_year', ascending=True)
)

#False 하면 내림차순
```

```python
# 여기서 title_year를 기준으로 중복되는 것을 제거해줍니다.
# drop_duplicates 는 겹치는 것이 등장하면 첫 등장을 제외하고 나머지를 지웁니다.
# subset 을 지정해, 어떤 컬럼을 기준으로 drop할 것인지 정합니다.
(
movie[['movie_title', 'title_year', 'imdb_score']]
.sort_values(['title_year', 'imdb_score'], ascending=False) #연도와 점수 정렬
.drop_duplicates(subset='title_year') #맨 첫 번째 행만 남기고 drop
)
```

```python
(
    movie[['movie_title', 'title_year', 'imdb_score']]
    .groupby('title_year', as_index=False)
    .apply(lambda df: df.sort_values('imdb_score', ascending=False).head(1))
    .droplevel(0)
    .sort_values('title_year', ascending=False)
)
```

```python
# 각 연도별, 등급별 최소 예산 영화를 찾습니다.

(
movie[['movie_title', 'title_year', 'content_rating', 'budget']]
.sort_values(['title_year', 'content_rating', 'budget'], ascending=[False, False, True]) # 내림 내림 오름!
.drop_duplicates(subset=['title_year', 'content_rating'])
)
```

```python
# 2. college 에서 UGDS 의 최대값을 구하는데 도시별로.
college.head()
#INSTNM, CITY, UGDS
```

```python
 college[['INSTNM', 'CITY', 'UGDS']].sort_values(['CITY', 'UGDS'], ascending=[True, False])
```

> city 가 같은 경우엔 ugds 순으로 배열

```python
(
    college[['INSTNM', 'CITY', 'UGDS']]
    .groupby(['CITY']) #여러 City에 대한 DF 생성
    .apply( # 모든 DF에 UGDS로 정렬 후 맨 위에 꺼만 솔트
        lambda df: df.sort_values('UGDS', ascending=False).head(1))
    .droplevel(1)
    .sort_index() #생성된 인덱스에 대해 정렬
)
```

```python
college2 = college[['INSTNM', 'STABBR', 'RELAFFIL','UGDS']]
```

```python
college2
```

```python

(
    college2.groupby(['STABBR','RELAFFIL'])
.apply(lambda df: df.nlargest(1,'UGDS'))
.droplevel(0)
.sort_index()
)
```
