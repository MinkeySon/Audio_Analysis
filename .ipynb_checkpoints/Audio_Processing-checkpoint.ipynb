# Mu-Law Encoding <br>
- 사람은 소리가 커질수록 차이 구별 x 작은 값에는 높은 분별력, 큰 값끼리는 낮은 분별력

```python
def mu_law(x, mu = 255):
    return np.sign(x) * np.log(1 + mu * np.abs(x)) / np.log(1+mu)
```

```python
x = np.linspace(-1,1,1000)
x_mu = mu_law(x)

plt.figure(figsize = [6,4])
plt.plot(x)
plt.plot(x_mu)
plt.show()
```

```python
wav_mulaw = mu_law(normed_wav)
ipd.Audio(wav_mulaw, rate = sr)
```

# Time-Frequency-Representation
- 모든 신호는 frequency(주파수), magnitude, phase가 다른 정현파(sinusolida signal) 조합으로 나타낼 수 있음
- 퓨리에 변환은 조합된 정현파의 합(하모니) 신호에서 그 신호를 구성하는 정현파들을 각각 분리해내는 방법

```python
def Sinusoid(A,f,phi,fs,t):
    t=np.arange(0,t,1.0/fs)
    x=A * np.cos(2*np.pi*f*t+phi)
    return x
```

# 푸리에 변환(Fourier Transform)
- 푸리에 변환은 임의의 입력 신호를 다양한 주파수를 갖는 주기 함수들의 합으로 분해하여 표현하는 것.

```python
# DFT 변환
def DFT(x):
    N = len(x)
    X = np.array([])
    nv = np.arrange(N)
    
    for k in range(N):
        s = np.exp(1j*2*np.pi*k/N*nv)
        X = np.append(X,sum(x*np.conjugate(s)))
```
